```python
import torch

# step 1
# input data X
X = torch.tensor([1.,2.,3.,4.])

# input y 
y = torch.tensor([2.,4.,6.,8.])

# step 2 -- wrong guesses
w = torch.tensor(0.0)
b = torch.tensor(0.0)

# step 3 
y_pred = w*X+b

loss = ((y_pred-y)**2).mean()

print(f"predicted : {y_pred}")
print(f"loss : {loss}")
print(f"Actual values : {y}")

```
