```python
import torch
# print(torch.__version__)

"""
creating tensors
"""
# seiries data or 1-D tensor
X1 = torch.tensor([1,2,3,4,5])
print(X1)


```

```python

# 2-D tensor

X2 = torch.tensor([[1,2],
                  [3,4]])
print(X2)
```

```python
# for initial values of tensor we can use torch.zeros, torch.ones, torch.rand

X3 = torch.zeros(3,2)
print(X3)
```

```python
# for initial values of tensor we can use torch.zeros, torch.ones, torch.rand

X4 = torch.ones(3,2)
print(X4)
```

```python
# for initial values of tensor we can use torch.zeros, torch.ones, torch.rand

X5 = torch.rand(3,2)
print(X5)
```

```python
# tensor shapes

X3.shape
```

```python
i = torch.tensor([10,20,30])
print(i[1])
```

```python
l = torch.tensor([[1,2,3],
                  [4,5,6]])
print(l[0,1])
print(l[1,2])
```

```python
a = torch.tensor([1,2,3])
b = torch.tensor([4,5,6])

print(a*b)
print(a+b)

print(a*3)
print((a+b)*4)
```

```python
import numpy as np
arr = np.array([1,2,3])
print(arr)

# creating tensors using numpy

n = torch.from_numpy(arr)
print(n)
```
